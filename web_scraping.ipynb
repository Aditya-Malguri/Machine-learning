```python
def scrape_page(page_number):
  url = Base.format(page_number)
  response = requests.get(url,headers=headers)
  soup = BeautifulSoup(response.text,'lxml')
  job_cards = soup.select("div.jobCardNova_bigCard__W2xn3")
  print(f"Found {len(job_cards)} jobs")

  jobs = []
  for card in job_cards:
    #title
    job_title = card.select_one('h3.jobCardNova_bigCardTopTitleHeading__Rj2sC a')
    title = job_title.text.strip() if job_title else None

    #company name
    comapny_name = card.select_one('span.jobCardNova_bigCardTopTitleName__M_W_m')
    company = comapny_name.text.strip() if comapny_name else None

    #exp
    experience_tags = card.select('span.jobCardNova_bigCardCenterListExp__KTSEc')
    exp = experience_tags[0].text.strip() if experience_tags else None

    #salary
    sal = None
    for   tag in experience_tags:
      if "LPA" in tag.text or "Lakh" in tag.text:
        sal = tag.text.strip()

    #location
    location_tag = card.select('div.jobCardNova_bigCardCenterListLoc__usiPB span')
    location = location_tag[0].text.strip() if location_tag else None

    skill_tag = card.select("ul.jobCardNova_skillsLists__7YifX li")
    skills = ", ".join([s.text.strip() for s in skill_tag]) if skill_tag else None

    # Posted date
    posted_tag = card.select_one("span.jobCardNova_postedData__LTERc")
    posted = posted_tag.text.strip() if posted_tag else None

    # Job URL
    url_tag = card.select_one("h3.jobCardNova_bigCardTopTitleHeading__Rj2sC a")
    job_url = "https://www.shine.com" + url_tag["href"] if url_tag else None


    jobs.append({
        "title":title,
        "company":company,
        "exp":exp,
        "salary":sal,
        "location":location,
        "skills":skills,
        "posted":posted,
        "job_url":job_url,
    })

  return jobs
```

```python
all_jobs = []
for page in range(1,1000):
  jobs = scrape_page(page)
  all_jobs.extend(jobs)
```

```python
df = pd.DataFrame(all_jobs)
```
